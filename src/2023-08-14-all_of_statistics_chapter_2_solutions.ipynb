## 1.

First, note that

$$
F(x)
= \mathbb{P}(X \leq x)
= \mathbb{P}(\{X = x\} \cup \{X < x\})
= \mathbb{P}(X = x) + \mathbb{P}(X < x).
$$

Moreover, by monotonicity,

$$
F(x-)
= \lim_{n \rightarrow \infty} F(x - 1 / n)
= \lim_{n \rightarrow \infty} \mathbb{P}(X \leq x - 1 / n)
= \mathbb{P}(X < x).
$$

Combining the above two equations yields the desired result.

## 2.

```python
x = np.arange(1, 7)
F = 1. / 10. * (x >= 2) + 1. / 10. * (x >= 3.) + 8 / 10. * (x >= 5)
plt.step(x, F, where="post")
plt.title("CDF of $X$");
```

Using $F$ we find that

$$
\mathbb{P}(2 < X \leq 4.8)
= F(4.8) - F(2)
= 0.2 - 0.1
= 0.1
$$

and

$$
\mathbb{P}(2 \leq X \leq 4.8)
= F(4.8) - F(2-)
= 0.2 - 0
= 0.2.
$$
